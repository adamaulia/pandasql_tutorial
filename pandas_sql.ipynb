# Data Aggregation 

```python
pysqldf("SELECT * FROM df_cars").head()
```

```python
# business question, what's the most frequest car by manufacture
qry_1 = """
select make, count(1)n from df_cars
group by 1
order by 2 desc
limit 10
"""
pysqldf(qry_1)
```

```python
# business question, what's average car price by manufacture
qry_2 = """
select make, avg(price)avg_price from df_cars
group by 1
order by 2 desc
limit 10
"""
pysqldf(qry_2)
```

```python
# business question, what's average car price, count by gear type and fuel type
qry_3 = """
select gear_type, fuel_type, count(1)n, avg(price)avg_price from df_cars
group by 1,2
order by 4 desc
limit 10
"""
pysqldf(qry_3)
```

# Row Number 

```python
# top 3 cars from each manufacture 
qry_4 = """  
select * from (
    select *, row_number() over(partition by make order by price desc) rank from df_cars
)x
where rank <= 3
"""
pysqldf(qry_4)
```

```python
qry_5 = """  
select * from (
    select *, row_number() over(partition by make order by price desc) rank from df_cars
)x
where rank = 1
"""

qry_6 = """  
select * from (
    select *, row_number() over(partition by make order by price desc) rank from df_cars
)x
where rank = 2
"""

df_cars_top_1 = pysqldf(qry_5)
df_cars_top_2 = pysqldf(qry_6)
```

# Join 

```python
qry_7 = """
select a.make, a.model as model_1 , a.version version_1 , a.price as price_1,
    b.model as model_2 , b.version version_2 , b.price as price_2
    from (
    select make, model, version, price, rank as rank_1 from df_cars_top_1
)a
left join (
    select make, model, version, price, rank as rank_2 from df_cars_top_2
)b
on  a.make = b.make

"""
pysqldf(qry_7)
```
